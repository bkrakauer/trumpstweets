```python
import pandas as pd
```

```python
df = pd.read_csv('tweetdb.csv')
```

```python
df.created_at = pd.to_datetime(df.created_at)
```

```python
freq_conversion = {"day": "d", "week": "w", "month": "m"}         #check this, of course
```

```python
index = pd.date_range(min(df.created_at), max(df.created_at), freq='d')
```

```python
index
```

```python
df["month"] = df.created_at.dt.month
```

```python
df["day"] = df.created_at.dt.day
```

```python
df["week"] = df.created_at.dt.week
```

```python
# fix month:

#df.month = df.month + 2
df.month = df.month.apply(lambda x: x - 13 if x > 10 else x)
df.month = df.month.apply(lambda x: x + 1 if x < 2 else x - 1)
```

```python
df.week = df.week.apply(lambda x: x - 47 if x > 47 else x - 22)
```

```python
# 11/25/2015 is the first day in our set!

def convert_day(row):
    day = 0
    
    if row.month == 1:  #that is, nov 2015!
        day = row.day - 24
    elif row.month == 2:
        day = row.day + 7 # if we're dec 1, 2015, that's the 8th day in the data set (counting from 1)
    elif row.month == 3:
        day = row.day + 38
    elif row.month == 4:
        day = row.day + 69
    elif row.month == 5:
        day = row.day + 98 # this accounts for leapday!
    elif row.month == 6:
        day = row.day + 129
    elif row.month == 7:
        day = row.day + 159
    elif row.month == 8:
        day = row.day + 190
    elif row.month == 9:
        day = row.day + 220
    return day    
```

```python
df.day = df.apply(convert_day, axis = 1)
```

```python
df.drop(["Unnamed: 0"], axis=1, inplace=True)
df
```

```python
df.to_csv("tweetdb.csv")
```
